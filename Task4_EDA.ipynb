# EDA of AMEO Data
## 1.1	Introduction
### 1.1.1	Dataset Description
The Aspiring Mind Employment Outcome 2015 (AMEO) dataset, released by Aspiring Minds, focuses on employment outcomes for engineering graduates. It includes dependent variables such as Salary, Job Titles, and Job Locations, along with standardized scores in cognitive skills, technical skills, and personality skills. With around 40 independent variables and 4000 data points, these variables encompass both continuous and categorical data. The dataset also includes demographic features and unique identifiers for each candidate.
## 1.1.2	Objective
The goal of this Exploratory Data Analysis (EDA) is to extensively investigate the provided dataset, with a particular emphasis on understanding the link between various variables and the target variable, Salary.
The key aims of this analysis include:  
1.	Providing a detailed explanation of the dataset’s features.      
2. Find any observable patterns or trends in the data.    
3. Investigating the relationships between the independent factors and the target variable(salary).      
4. Identify any outliers or abnormalities in the dataset.    
5.	Offering practical insights and recommendations based on the analysis.

```python
df = data.drop(columns =['Unnamed: 0', 'ID', 'CollegeID', 'CollegeCityID'])
#df.head()
```

**Dropping the rows where DOL is less then DOJ**

```python
# DataFrame containing the removed rows
removed_rows = df[df['DOL'] < df['DOJ']]
print(removed_rows.shape)

# Remove rows where DateOfLeaving < DateOfJoining
df = df[df['DOL'] >= df['DOJ']]
print(df.shape)
```

## Data Conversion


```python
from datetime import datetime

today_date = datetime.today().strftime('%Y-%m-%d')
df['DOL'] = df['DOL'].replace('present', today_date)

df['DOJ'] = pd.to_datetime(df['DOJ'])
df['DOB'] = pd.to_datetime(df['DOB'])
#print(df.head())
```

**Replacing 'present' value to today_date of 'DOL' variable**

```python
from datetime import datetime
today_date = datetime.today().strftime('%Y-%m-%d')
df['DOL'] = df['DOL'].replace('present', today_date)
#print(df.head())
```

## Feature Engineering

```python
df['Age'] = 2024 - df['DOB'].dt.year
#print(df.head())
```

# EDA (Exploratory Data Analysis)

### Categorical Data

```python
# Select categorical columns excluding date columns
categorical_columns = df.select_dtypes(include=['object', 'category'], exclude=['int64', 'float64', 'datetime64[ns]'])

# Create a new DataFrame with categorical columns
categorical_df = pd.DataFrame(categorical_columns)
categorical_df = categorical_df.drop('DOL', axis=1)
print(categorical_df.head())
```

**'Others' are the values where '0' or '-1' was entered**

```python
import numpy as np
for column in categorical_df:
    if categorical_df[column].dtype == 'object':  # Check if the column contains object (string) values
        categorical_df[column] = categorical_df[column].apply(lambda x: 'Others' if x == '0' or x == '-1' else x)
```

```python
for column in categorical_df:
    if categorical_df[column].dtype == 'object':  # Check if the column contains string values
        categorical_df[column] = categorical_df[column].str.lower()
```

#### **Non-visual Analysis**

##### **count, unique, unique & value_counts**

```python
def unique_nunique(df):
    for column in categorical_df:
    unique_values = df[column].unique()
    #value_counts = df[column].value_counts()
    column_unique_values = df[column].nunique()
    print('*No. of unique values in', column, ':-', column_unique_values)
    print('\n*Unique values in', column, ':-', unique_values, '\n','=='*50)

categorical_n_unique = unique_nunique(categorical_df)
print(categorical_n_unique)
```

```python
def value_counts(df):
    for column in categorical_df:
    value_counts = categorical_df[column].value_counts()
    print('*Value counts in', column, ':-\n')
    print(value_counts[:12])
    print('='*100,'\n')

categorical_value_counts = value_counts(categorical_df)
print(categorical_value_counts)
```

##### **Salary**

```python
plt.figure(figsize=(5, 4))
sns.boxplot(data=numerical_df, y='Salary', color='skyblue')
plt.title('Box Plot of Salary')
plt.ylabel('Salary')
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import norm, skew

salary_values = numerical_df['Salary']

# Calculate mean, median, mode, and skewness
mean_salary = salary_values.mean()
median_salary = salary_values.median()
mode_salary = salary_values.mode()[0]
skewness_salary = skew(salary_values)

# Plot histogram
plt.figure(figsize=(6, 5))
sns.histplot(salary_values, kde=True, bins=20, label = f"Skewness : {round(numerical_df['Salary'].skew(),2)}", color='skyblue')

# Add mean, median, mode, and skewness information
plt.axvline(mean_salary, color='red', linestyle='--', label=f'Mean: {mean_salary:.2f}')
plt.axvline(median_salary, color='green', linestyle='--', label=f'Median: {median_salary:.2f}')
plt.axvline(mode_salary, color='purple', linestyle='--', label=f'Mode: {mode_salary:.2f}')

# Add distribution line
x = np.linspace(salary_values.min(), salary_values.max(), 100)
#plt.plot(x, norm.pdf(x, mean_salary, salary_values.std()), color='orange', label='Distribution')

plt.title('Histogram of Salary with Mean, Median, Mode, and Skewness')
plt.xlabel('Salary')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

###### **Observation:**
1. Salary variations are significant throughout the sample.
2. With a skewness value of roughly 6, the data shows strong positive skewness, which deviates from a normal distribution.
3. The mean, median, and mode—three measures of central tendency—are about equal.

```python
import random
n = 30
salary_random = random.sample(df1['Salary'].tolist(),n)
print(salary_random)
```

**Sample values calculations**

```python
from scipy.stats import t,norm
import statistics
print('Sample Mean: ', statistics.mean(salary_random))
print('Sample Standard Deviation: ', statistics.stdev(salary_random))
```

```python
sample_size = n
sample_mean = statistics.mean(salary_random)
pop_mean = data.Salary.mean()
sample_std = statistics.stdev(salary_random)
pop_mean
```

```python
def t_score(sample_size, sample_mean, pop_mean, sample_std):
  numerator = sample_mean - pop_mean
  denomenator = sample_std / sample_size**0.5
  return numerator / denomenator
```

**Calculating t-value**

```python
t_value = t_score(sample_size, sample_mean, pop_mean, sample_std)
print("t-value :",t_value)
```

**Calculating t-statistic**

```python
confidence_level = 0.95
alpha = 1 - confidence_level
t_critical = t.ppf(1 - alpha/2, df = 99)
print("t-critical:\t",t_critical)
```

```python
if (t_value < t_critical):
  print("There is not enough evidence to reject the Null Hypothesis")
else:
  print("There is sufficent evidence to reject the Null Hypothesis")
```

```python
p_value = 2 * (1.0 - norm.cdf(np.abs(t_value)))
print("p_value :\t ", p_value)
if(p_value > alpha):
  print("There is not enough evidence to reject the Null Hypothesis")
else:
  print("There is sufficent evidence to reject the Null Hypothesis")
```
